```python
df = pd.read_csv('/content/datasettapriorifixxxx.csv')
df.head()
```

```python
df.drop(["address","district","lat","long","Cluster"], axis="columns", inplace=True)
df.columns = ["day","time","accident_type","age","gender","suspect_vehicle"]
print(df.shape)
df.head()
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Membaca data dari file CSV
# df = pd.read_csv('nama_file.csv')

# Mengonversi data menjadi list of lists (transaksi)
transactions = df.values.tolist()

# Menggunakan TransactionEncoder untuk mengonversi data transaksi menjadi matriks biner
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_transformed = pd.DataFrame(te_ary, columns=te.columns_)

# Menjalankan algoritma Apriori untuk mendapatkan itemset yang sering muncul
frequent_itemsets = apriori(df_transformed, min_support=0.2, use_colnames=True)

# Menampilkan itemset yang sering muncul
print("Frequent Itemsets:")
print(frequent_itemsets)

# Menjalankan algoritma Association Rules untuk mendapatkan aturan asosiasi
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

# Menampilkan aturan asosiasi
print("\nAssociation Rules:")
print(rules)
```
