```python
!pip install apyori

import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv('/content/datasetApriorinew.csv')
df.head()
```

```python
df.drop(["address","district","lat","long","Cluster"], axis="columns", inplace=True)
df.columns = ["day","time","accident_type","age","gender","suspect_vehicle"]
print(df.shape)
df.head()
```

```python
records = []
for i in range(0,1338):
  records.append([str(df.values[i,j]) for j in range(0,6)])

records
```

```python
association_rule = apriori(records, min_support=0.05, min_confidence=0.2, min_lift=3, min_length=2)
```

```python
assoc_rule=list(association_rule)
assoc_rule
```

```python
for item in assoc_rule:
  pair = item[0]
  items = [x for x in pair]
  print(items)
  print("support: "+str(item[1]))

  print("confidance: "+str(item[2][0][2]))
  print("lift: "+str(item[2][0][3]))
  print("=========================================================")
```
