```python
df = pd.read_csv('/content/DATAAPRIORI.csv')
df.drop(["NO", "TGL"], axis="columns", inplace=True)
df.columns = ["day", "time", "address", "district", "accident_types", "age","gender","suspect_vehicle", "victim_vehicle"]
print("Columns after drop:", df.columns)
df.head()
```

```python
df["address"] = df["address"].map(lambda x: str(x).strip().lower())
df["district"] = df["district"].map(lambda x: str(x).strip().lower())
df.head()
```

```python
df["time"] = pd.to_datetime(df['time'], format='%H:%M', errors='coerce')
df["time"] = pd.to_datetime(df["time"]).dt.hour
```

```python
def conv_time(hour):
    if 5<=hour<=11:
        return "Morning"
    elif 12<=hour<=17:
        return "Afternoon"
    elif 18<=hour<=22:
        return "Evening"
    else:
        return "Night"

df["time"] = df["time"].map(lambda x: conv_time(x))
df.head()
```
